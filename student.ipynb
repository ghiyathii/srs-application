# **Quick Sort**

```python
# Python quick sort algorithm

def quick_sort(array):

#  is the start of a Python function definition for a sorting algorithm

    if len(array) <= 1:

# checks if the length of the students array is less than or equal to 1

        return array

# If it is, then the function returns the array as is

    pivot = array[len(array) // 2][1]

# Here I selected pivot 0
# Divide the three lines of the "array" into three 
# sub-array: "left", "middle", and "right".

    left = [x for x in array if x[1] < pivot]


# Create a list of all the elements in the array whose first element 
                                            #is less than the pivot
                                                          

    middle = [x for x in array if x[1] == pivot]

# Create a list of all the elements in the array whose first element is
                                                   # equal to the pivot
   
    right = [x for x in array if x[1] > pivot]

# Create a list of all the elements in the array whose first element is
                                               # greater than the pivot

    return quick_sort(left) + middle + quick_sort(right)

# Calling the 'Quick Sort' function on the 'right' sub-array
```

```python
Students = quick_sort(Students)

# The list of students is called after the sorted
```

# **sequential_search**

```python
# Python sequential search algorithm

def sequential_search(array, target):

# (array) is the array that I would like to search in it
# (target) is the target value that I want to search for in the (array)

    for i in range(len(array)):

# variable "i" is used as a counter in the for loop
# "i" take all of the numder from 0 to -1


        if array[i][1] == target:
# Check if the second element of the current element in the "array" is 
                                   #equal to the value of the "target"

            return i

# If equal to "target", the function returns the value of i
# i is the index of the target

    return -1
```

```python
sequential_search(Students, 'Majed')
```

# **binarySearch**

```python
# Python binarySearch algorithm

def binarySearch(array, l, r, x):
  
# l = the right index
# r = the left index
# x = is the target


    if r >= l:

# check if (r) is greater than or equal to the left index (l) 

        mid = (l + r) // 2

        if array[mid][0] == x:

# Check if [0] equals x

            return mid

# If this condition is true, return the middle of the index

        elif array[mid][0] > x:

          
# If true, it means that the target value x is in the left half

            return binarySearch(array, l, mid-1, x)


# recursive call to the binarySearch function


        else:
            return binarySearch(array, mid + 1, r, x)
 
    else:
        return -1
```
